```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

titanic_df = pd.read_csv('titanic/titanic_train.csv')
titanic_df.head(3)
```

- Passengerid : 탑승자 일련번호
- survived :  생존 여부(0 : 사망, 1 : 생존)
- Pclass : 티켓의 선실 등급(1 : 일등석, 2 : 이등석, 3 : 삼등석)
- name : 탑승자 이름
- sex : 탑승자 성별
- Age : 탑승자 나이
- sibsp : 같이 탑승한 형제자매 또는 배우자 인원수
- parch : 같이 탑승한 부모님 또는 어린이 인원수
- ticket : 티켓 번호
- fare : 요금
- cabin : 선실 번호
- embarked : 중간 정착 항구(C : Cherbourg, Q : Queenstown, S : Southhampton)

```python
print('\n ### train 데이터 정보 ### \n')
print(titanic_df.info())
```

**NULL 컬럼들에 대한 처리**

```python
titanic_df['Age'].fillna(titanic_df['Age'].mean(),inplace = True) # 결측값을 mean값으로 대체
titanic_df['Cabin'].fillna('N', inplace = True) # 결측값을 'N'으로 대체
titanic_df['Embarked'].fillna('N', inplace = True) # 결측값을 'N'으로 대체

print('데이터 세트 Null 값 개수 :', titanic_df.isnull().sum().sum())
```

```python
print('데이터 세트 Null 값 개수 :', titanic_df.isnull().sum())
```

```python
print('Sex값 분포 : \n', titanic_df['Sex'].value_counts(),'\n')
print('Cabin값 분포 : \n', titanic_df['Cabin'].value_counts(),'\n')
print('Embarked값 분포 : \n', titanic_df['Embarked'].value_counts())
```

```python
titanic_df['Cabin'] = titanic_df['Cabin'].str[:1] # 번호를 제외한 알파벳만 추출
print(titanic_df['Cabin'].head(3))
```

```python
titanic_df.groupby(['Sex', 'Survived'])['Survived'].count()
```
